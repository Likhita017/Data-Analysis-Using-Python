```python
#Take salary from the user.If the salary is more than 50k there is a 10% tax.Print his final salary.
s=int(input("Enter salary"))
if(s>=50000):
  print("Your final salary is",s-((0.1)*s))
else:
  print("There is no tax deduction")
```

```python
#Get the salary  from the user if the salary is >= 50k apply 10% tax deduction .if the salary is < 50k apply 5% tax,if the salary > 70kapply 12% tax if salary exceeds 1L print you are promoted
s=int(input("Enter salary"))
if(s<50000):
  print("Your final salary is",s-((0.05)*s))
elif(s<=70000):
  print("Your final salary is",s-((0.1)*s))
elif(s<=100000):
  print("Your final salary is",s-((0.12)*s))
else:
  print("You are promoted")
```
