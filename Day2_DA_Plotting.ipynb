```python
#There are 5 subjects psychology,nanotechnology,medicine,onchology,anthropology. Two students have applied to do masters in any of these five fields with a line plot show in which they can exel.The cutoff for each subject  is 70.
#s1=[67,45,73,71,57]
#s2=[43,67,68,53,85]
sub=['psychology','nanotechnology','medicine','oncology','anthropology']
s1=[67,45,73,71,57]
s2=[43,67,68,53,85]
cut=[70,70,70,70,70]
#syntax- plt.plot(X,Y,label,comor,marker,linewidth,linestyle)
#student 1
plt.scatter(sub,s1,label="Student1",color='m',marker="P")
#student 2
plt.scatter(sub,s2,label="Student2",color='y',marker="p")
#cutoff
plt.plot(sub,cut,label="cutoff",color='k',linewidth=1,linestyle=":")
plt.xlabel("subjects")
plt.ylabel("marks")
plt.legend()
```

```python
#An India vs australia is on live. India has given 298 as for australia.The runs for every  5 overs is given as follows.
#India score: 25,51,84,131,160,189,220,250,267,297 Australia score has completed only 25 overs.
#Australia: 15,41,94,110,151
#Plot overs vs runs for both the countries with india as a blue line and australia as yellow line.
#Give appropriate labels and legend the marker for india is flag & australia is plus
#Take overs from aarange func create runs as arrays
import numpy as np
io=np.arange(5,51,5)
ao=np.arange(5,26,5)
ir=np.array([25,51,84,131,160,189,220,250,267,297])
ar=np.array([15,41,94,110,151])
plt.title("India vs Australia")
plt.plot(io,ir,label='India runs',color='b',marker='>')
plt.plot(ao,ar,label='Australia runs',color='y',marker='p')
plt.xlabel('Overs')
plt.ylabel('Runs')
plt.legend()
```
